# Exercice 2.3-Python

On desire résoudre les deux équations différentielles ci-dessous:

$\frac{dx}{dt} = 0.02\cdot x - 0.00004\cdot x\cdot y$ avec $x_0$ = 500

$\frac{dy}{dt} = 0.0004\cdot x\cdot y - 0.04\cdot y$ avec $y_0$ = 200

## Solution

Description du problème

```python
#ATTENTION: importer le solveur
from scipy.integrate import odeint
import numpy as np
```

```python
# Constantes
A = 0.02
B = 0.00004
C = 0.0004
D = 0.04

# Conditions initiales
x0 = 500
y0 = 200
```

Définition des équations différentielles

```python
def equations(variables, t, A, B, C, D):
    x, y = variables
    
    # Mes équations différentielles
    dxdt = A * x - B * x * y
    dydt = C * x * y - D * y
    
    return [dxdt, dydt]
```

Résoudre le problème à l'aide du solveur d'équations différentielles

```python
# Définir les conditions initiale
variables0 = [x0, y0]
t = np.linspace(0, 1000, 1001)

print(t)
```

```python
valeurs = odeint(equations, variables0, t, (A, B, C, D))
```

```python
print(valeurs)
```

Une fois qu'on a résolu le problème, on peut afficher les variations de x et y sur un graphique

```python
# On importe la bibliothèque utilisée pour la visualisation (pylab)
import matplotlib.pylab as plt
```

```python
plt.plot(t, valeurs)
plt.xlabel('t (jours)')
plt.ylabel('x, y')
plt.legend(['x', 'y'], loc='best')
plt.show()
```
