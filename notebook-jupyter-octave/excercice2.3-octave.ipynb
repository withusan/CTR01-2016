# Exercice 2.3-Matlab

On desire résoudre les deux équations différentielles ci-dessous:

$\frac{dx}{dt} = 0.02\cdot x - 0.00004\cdot x\cdot y$ avec $x_0$ = 500

$\frac{dy}{dt} = 0.0004\cdot x\cdot y - 0.04\cdot y$ avec $y_0$ = 200

## Solution

Description du problème

```octave_kernel
% uniquement pour octave
pkg load odepkg % utilisation du solveur ode45 de matlab
```

```octave_kernel
% Constantes
A = 0.02;
B = 0.00004;
C = 0.0004;
D = 0.04;

% Conditions initiales
x0 = 500;
y0 = 200;
```

Définition des équations différentielles

```octave_kernel
function dvariables = equations(t, variables, A, B, C, D)
    x = variables(1);
    y = variables(2);
    
    # Mes équations différentielles
    dxdt = A * x - B * x * y;
    dydt = C * x * y - D * y;
    
    dvariables = [dxdt; dydt];
end
```

Résoudre le problème à l'aide du solveur d'équations différentielles

```octave_kernel
# Définir les conditions initiale
variables0 = [x0, y0];
tspan = 0 : 1 : 1000;
```

```octave_kernel
[t, valeurs] = ode45(@(t, variables) equations(t, variables, A, B, C, D), tspan, variables0);
```

```octave_kernel
plot(t, valeurs);
h = legend('x', 'y');
set(h, 'fontsize', 6);
set(gca, 'fontsize', 6);
xlabel('t (jours)');
ylabel('x, y');
```
